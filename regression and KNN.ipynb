#### (c)

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
x=np.random.uniform(0,1,100)
y=np.random.uniform(x+np.cos(2*np.pi*x)-0.1,x+np.cos(2*np.pi*x)+0.1)
D=pd.DataFrame({"x":x,"y":y})
model = LinearRegression()
model.fit(x.reshape(-1,1), y)
intercept = model.intercept_
slope = model.coef_[0]
y1=model.predict(x.reshape(-1,1))
x_sorted = np.sort(x)
true_y_sorted = x_sorted + np.cos(2 * np.pi * x_sorted)
plt.scatter(D["x"],D["y"],label="scatter plot")
plt.plot(x,y1,color="red",label="predicted linear model")
plt.plot(x_sorted, true_y_sorted, color="green", linestyle="dashed", label="True Regression Model")

plt.xlabel("x")
plt.ylabel("y")
plt.title("scatterplot of x and y with predicted and true linear regression in uniform distribution")
plt.legend()
plt.show()
model.predict(np.array([[0.5]]))[0]
```

From the result it seems that when x=0.5, the predicted y is around 0.569, so $\hat f(0.5;D)=0.569$

#### (d)

```python
preds=list()
for i in range(500):
    x = np.random.uniform(0, 1, 100)
    y = np.random.uniform(x + np.cos(2 * np.pi * x) - 0.1, x + np.cos(2 * np.pi * x) + 0.1)
    model = LinearRegression()
    model.fit(x.reshape(-1, 1), y)
    y_pred = model.predict(np.array([[0.5]]))[0]
    preds.append(y_pred)
preds= np.array(preds)
plt.hist(preds)
plt.show()
```

#### (a)

```python
import pandas as pd
import seaborn as sns
smarket=pd.read_csv("C:\\Users\\DELL\\Downloads\\smarket.csv")
print(smarket.describe())
sns.countplot(x="Direction", data=smarket)
```

From the numerical summearies it seems that the mean value of lags have a decreasing trend with the increase of days, for example, the mean of lag5 is smaller than that of lag4. Such trend is contrary to that of standard deviation, where the increase of lag days leads to the increase of std. The minimum, median, maximum and first and third quantile for all lag days are similar to each other, some of them are even in same number. The standard deviaton of volumn is small, indicating there doesn't exist much variation of it. Although the mean value of today's return is positive, there exist lots of number that are negative, which can be viewed from the minimum and 25% quantile.
From the figure summaries it seems that the number of ups for direction is bigger than that of downs but doesn't exceed much.

#### (b)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, accuracy_score
train=smarket[smarket["Year"]<2005]
test=smarket[smarket["Year"]==2005]
X_train=train.loc[:,["Lag1","Lag2"]]
y_train=train.loc[:,["Direction"]]
X_test=test.loc[:,["Lag1","Lag2"]]
y_test=test.loc[:,["Direction"]]
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred = lda.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
accuracy = accuracy_score(y_test, y_pred)
print(f"Overall Accuracy: {accuracy:.4f}")
```

#### (c)

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, accuracy_score
train=smarket[smarket["Year"]<2005]
test=smarket[smarket["Year"]==2005]
X_train=train.loc[:,["Lag1","Lag2"]]
y_train=train.loc[:,["Direction"]]
X_test=test.loc[:,["Lag1","Lag2"]]
y_test=test.loc[:,["Direction"]]
qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train, y_train)
y_pred = qda.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
accuracy = accuracy_score(y_test, y_pred)
print(f"Overall Accuracy: {accuracy:.4f}")
```

#### (a)

```python
auto=pd.read_csv("C:\\Users\\DELL\\Downloads\\auto.csv")
auto["mpg01"]=pd.cut(auto["mpg"],bins=[-np.inf,25,np.inf],labels=[0,1])
```

#### (b)

```python
fig, axes = plt.subplots(3, 3, figsize=(18, 10))
features = ["displacement", "horsepower", "weight", "acceleration", "cylinders","year","origin"]
for i, feature in enumerate(features):
    sns.boxplot(x=auto["mpg01"], y=auto[feature], ax=axes[i//3, i%3])
    axes[i//3, i%3].set_title(f"{feature} vs mpg01")
plt.tight_layout()
plt.show()
```

From the plots it seems that the majority of variables do have a significant influence on mpg01 as their boxes of 0 and 1 don't overlap each other. From my perspeitive, displacement, horsepower, weight and cylinders are most useful for predicting mpg01 as the box for 0 and 1 in their boxplot don't overlap each other, and for origin the box for 1 nearly take up the whole column, which is not good for boxplot.

#### (c)

```python
from sklearn.model_selection import train_test_split
X = auto.drop(columns=["mpg01"]) 
y = auto["mpg01"] 
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=123, stratify=y)
(y_train == 1).sum()
```

#### (d)

```python
X_train=X_train.loc[:,["displacement", "horsepower", "weight","cylinders"]]
X_test=X_test.loc[:,["displacement", "horsepower", "weight","cylinders"]]
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
y_pred = lda.predict(X_test)
y_pred1=lda.predict(X_train)
train_error = 1 - accuracy_score(y_train, y_pred1)
test_error = 1 - accuracy_score(y_test, y_pred)
print(f"Training Misclassification Rate: {train_error:.4f}")
print(f"Test Misclassification Rate: {test_error:.4f}")
var1, var2 = "horsepower", "weight"
y_train_np = np.array(y_train).astype(int)
y_pred_train_np = np.array(y_pred1).astype(int)
scatter = sns.scatterplot(x=X_train[var1], y=X_train[var2],hue=y_train_np, style=y_pred_train_np,palette={0: "blue", 1: "red"}, markers={0: "X", 1: "o"})
plt.xlabel(var1)
plt.ylabel(var2)
plt.title(f"Scatter Plot of {var1} vs {var2} (LDA Predictions)")
plt.legend(title="True mpg01")
plt.show()
```

#### (e)

```python
X_train=X_train.loc[:,["displacement", "horsepower", "weight","cylinders"]]
X_test=X_test.loc[:,["displacement", "horsepower", "weight","cylinders"]]
qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train, y_train)
y_pred = qda.predict(X_test)
y_pred1=qda.predict(X_train)
train_error = 1 - accuracy_score(y_train, y_pred1)
test_error = 1 - accuracy_score(y_test, y_pred)
print(f"Training Misclassification Rate: {train_error:.4f}")
print(f"Test Misclassification Rate: {test_error:.4f}")
var1, var2 = "horsepower", "weight"
y_train_np = np.array(y_train).astype(int)
y_pred_train_np = np.array(y_pred1).astype(int)
scatter = sns.scatterplot(x=X_train[var1], y=X_train[var2],hue=y_train_np, style=y_pred_train_np,palette={0: "blue", 1: "red"}, markers={0: "X", 1: "o"})
plt.xlabel(var1)
plt.ylabel(var2)
plt.title(f"Scatter Plot of {var1} vs {var2} (QDA Predictions)")
plt.legend(title="True mpg01")
plt.show()
```

#### (f)

From the result it seems that QDA is better than LDA, indicating that each class has their own covariance matrix.

#### (g)

```python
from sklearn.linear_model import LogisticRegression
X_train=X_train.loc[:,["displacement", "horsepower", "weight","cylinders"]]
X_test=X_test.loc[:,["displacement", "horsepower", "weight","cylinders"]]
reg=LogisticRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
y_pred1=reg.predict(X_train)
train_error = 1 - accuracy_score(y_train, y_pred1)
test_error = 1 - accuracy_score(y_test, y_pred)
print(f"Training Misclassification Rate: {train_error:.4f}")
print(f"Test Misclassification Rate: {test_error:.4f}")
var1, var2 = "horsepower", "weight"
y_train_np = np.array(y_train).astype(int)
y_pred_train_np = np.array(y_pred1).astype(int)
scatter = sns.scatterplot(x=X_train[var1], y=X_train[var2],hue=y_train_np, style=y_pred_train_np,palette={0: "blue", 1: "red"}, markers={0: "X", 1: "o"})
plt.xlabel(var1)
plt.ylabel(var2)
plt.title(f"Scatter Plot of {var1} vs {var2} (QDA Predictions)")
plt.legend(title="True mpg01")
plt.show()
```

#### (h)

```python
medians = X_train.median()
median_data_point = medians.values.reshape(1, -1) 
probability = reg.predict_proba(median_data_point)[:, 1] 
print(f"{probability[0]:.4f}")
```

#### (i)

```python
from sklearn.neighbors import KNeighborsClassifier
train_errors = []
test_errors = []
k_values = np.arange(1, 31)
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_train_pred = knn.predict(X_train)
    y_test_pred = knn.predict(X_test)
    train_error = 1- accuracy_score(y_train, y_train_pred)
    test_error = 1- accuracy_score(y_test, y_test_pred)
    train_errors.append(train_error)
    test_errors.append(test_error)
plt.figure(figsize=(10, 6))
plt.plot(k_values, train_errors, label='Training Error', color='blue', marker='o')
plt.plot(k_values, test_errors, label='Test Error', color='red', marker='x')
plt.show()
best_k_train = k_values[np.argmin(train_errors)]
best_k_test = k_values[np.argmin(test_errors)]
print(f"Best K based on Training Error: {best_k_train}")
print(f"Best K based on Test Error: {best_k_test}")
```

The best K for training error is 1, the best one for test is 23. I choose K=23 as training set is often more important.

#### (j)

```python
optimal_k =23
knn = KNeighborsClassifier(n_neighbors=optimal_k)
knn.fit(X_train, y_train)
y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)
train_error = 1 - accuracy_score(y_train, y_train_pred)
test_error = 1 - accuracy_score(y_test, y_test_pred)
print(f"Training Misclassification Rate: {train_error:.4f}")
print(f"Test Misclassification Rate: {test_error:.4f}")
var1, var2 = "horsepower", "weight"
y_train_np = np.array(y_train)
y_pred_train_np = np.array(y_train_pred)
scatter = sns.scatterplot(x=X_train[var1], y=X_train[var2], hue=y_train_np, style=y_pred_train_np, palette={0: "blue", 1: "red"}, markers={0: "X", 1: "o"})
plt.xlabel(var1)
plt.ylabel(var2)
plt.title(f"Scatter Plot of {var1} vs {var2} (KNN Predictions)")
plt.legend(title="True mpg01")
plt.show()
```

#### (k)

```python
median_values = X_train.median()
new_data_point = median_values.values.reshape(1, -1) 
probability = knn.predict_proba(new_data_point)
probability_mpg_above_25 = probability[0, 1]
print(f"Estimated probability of mpg > 25: {probability_mpg_above_25:.4f}")
```

The performance of logistic regression is the best, indicating that the relationships between the predictors and the outcome (mpg01) are better modeled using a logistic curve. It has a linear boundary.
